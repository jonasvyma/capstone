```python
# =============================
# 0 LIBRARIES & DATA LOAD
# =============================
import pandas as pd
import numpy as np
from datetime import datetime
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from prophet import Prophet
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("final_dataset.csv")

# Ensure proper date type
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
df = df.dropna(subset=["Date"])

print(f" Dataset loaded with {len(df)} rows and {df['Card Name'].nunique()} unique cards.")
```

```python
# Choose which price metric you want to test first
target_col = "Graded"  # or "New", "Used", "Graded"

# Prepare results container
results = []

for card in df["Card Name"].unique():
    sub = df[df["Card Name"] == card].copy()

    if target_col not in sub.columns or sub[target_col].isna().sum() > len(sub) * 0.5:
        continue  # skip if too many NaNs

    # Drop missing
    sub = sub.dropna(subset=[target_col])
    if len(sub) < 5:
        continue

    # Numeric time variable (ordinal)
    sub["t"] = (sub["Date"] - sub["Date"].min()).dt.days.values.reshape(-1, 1)

    # Regression
    X = sub["t"].values.reshape(-1, 1)
    y = sub[target_col].values.reshape(-1, 1)

    model = LinearRegression().fit(X, y)
    y_pred = model.predict(X)
    r2 = r2_score(y, y_pred)

    results.append({
        "Card Name": card,
        "R2": r2,
        "Coeff": model.coef_[0][0],
        "Intercept": model.intercept_[0]
    })

# Convert to DataFrame
reg_summary = pd.DataFrame(results).sort_values("R2", ascending=False)
print("Top cards by regression fit:")
print(reg_summary.head(10))
```

```python
from openpyxl import Workbook

# Create Excel writer
writer = pd.ExcelWriter("forecast_results.xlsx", engine="openpyxl")

price_columns = ["New", "Used", "Graded", "Boxonly"]  # adapt to your real names

for card in df["Card Name"].unique():
    sub = df[df["Card Name"] == card].copy()
    sub = sub.sort_values("Date")

    for col in price_columns:
        if col not in sub.columns:
            continue

        prophet_df = sub[["Date", col]].dropna()
        if len(prophet_df) < 10:
            continue

        prophet_df = prophet_df.rename(columns={"Date": "ds", col: "y"})

        # Prophet model
        model = Prophet()
        model.fit(prophet_df)

        # Forecast 12 months ahead (365 days)
        future = model.make_future_dataframe(periods=365, freq="D")
        forecast = model.predict(future)

        # Keep relevant columns
        result = forecast[["ds", "yhat", "yhat_lower", "yhat_upper"]]
        result["Card Name"] = card
        result["Price Type"] = col

        # Save to Excel sheet
        sheet_name = f"{card[:20]}_{col[:10]}"
        result.to_excel(writer, sheet_name=sheet_name, index=False)

writer.close()
print(" Prophet forecasts saved to forecast_results.xlsx")
```

```python
# ================================================
#  CLUSTERING BASED ON GRADED PRICE BEHAVIOUR
# ================================================
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Assume df already loaded from 'final_dataset.csv'
# and 'Date' is already converted to datetime

features = []

for card in df["Card Name"].unique():
    sub = df[df["Card Name"] == card].copy()

    # Skip cards with too many missing graded prices
    if "Graded" not in sub.columns or sub["Graded"].isna().sum() > len(sub) * 0.5:
        continue

    sub = sub.dropna(subset=["Graded"])
    if len(sub) < 5:
        continue

    # Compute key time-series statistics
    mean_price = sub["Graded"].mean()
    std_price = sub["Graded"].std()
    pct_change = sub["Graded"].pct_change().dropna()
    volatility = pct_change.std()
    growth = (sub["Graded"].iloc[-1] - sub["Graded"].iloc[0]) / sub["Graded"].iloc[0]

    features.append({
        "Card Name": card,
        "Mean": mean_price,
        "Std": std_price,
        "Volatility": volatility,
        "Growth": growth
    })

# Convert to DataFrame
feat_df = pd.DataFrame(features).dropna()

# Scale features
scaler = StandardScaler()
X = scaler.fit_transform(feat_df[["Mean", "Std", "Volatility", "Growth"]])

# Cluster with KMeans (try 35 clusters)
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
feat_df["Cluster"] = kmeans.fit_predict(X)

# Display results
print(" Cluster distribution (by graded price):")
print(feat_df["Cluster"].value_counts())

# Plot
plt.figure(figsize=(9,6))
sns.scatterplot(
    data=feat_df,
    x="Growth",
    y="Volatility",
    hue="Cluster",
    palette="viridis",
    s=120
)

# Add card name labels
for i, row in feat_df.iterrows():
    plt.text(
        row["Growth"] + 0.02,
        row["Volatility"],
        row["Card Name"],
        fontsize=8
    )

plt.title("Card Clusters  Graded Price: Growth vs Volatility")
plt.xlabel("Growth (Final vs Initial %)")
plt.ylabel("Volatility (std of returns)")
plt.legend(title="Cluster", loc="best")
plt.show()

# Save the feature dataset for reference
feat_df.to_csv("graded_clusters.csv", index=False)
print(" Saved: graded_clusters.csv")

```

```python
# ==============================================
#  CHARIZARD MARKET STUDY  Pokmon Collectibles
# ==============================================
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Optional: forecasting
from prophet import Prophet

# Load your full dataset
df = pd.read_csv("final_dataset.csv")
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

# ----------------------------------------------
# 1 Filter only Charizard cards
# ----------------------------------------------
char_df = df[df["Card Name"].str.contains("Charizard", case=False, na=False)].copy()

print(f"Found {char_df['Card Name'].nunique()} unique Charizard cards:")
print(char_df['Card Name'].unique())

# ----------------------------------------------
# 2 Basic descriptive statistics
# ----------------------------------------------
summary = (
    char_df.groupby("Card Name")["Graded"]
    .agg(["count", "mean", "std", "min", "max"])
    .sort_values("mean", ascending=False)
)
summary["volatility_pct"] = (summary["std"] / summary["mean"]) * 100
print("\n Charizard price summary:")
print(summary.round(2))

# ----------------------------------------------
# 3 Time-series comparison
# ----------------------------------------------
plt.figure(figsize=(12,6))
sns.lineplot(data=char_df, x="Date", y="Graded", hue="Card Name", linewidth=2)
plt.title(" Charizard Cards  Graded Price Evolution")
plt.xlabel("Date")
plt.ylabel("Graded Price ()")
plt.grid(alpha=0.3)
plt.legend(title="Charizard Variant", fontsize=8)
plt.show()

# ----------------------------------------------
# 4 Normalized relative performance
# ----------------------------------------------
char_norm = char_df.copy()
char_norm["Norm_Graded"] = char_norm.groupby("Card Name")["Graded"].transform(lambda x: x / x.iloc[0] * 100)

plt.figure(figsize=(12,6))
sns.lineplot(data=char_norm, x="Date", y="Norm_Graded", hue="Card Name", linewidth=2)
plt.title(" Normalized Charizard Index (Base = 100)")
plt.xlabel("Date")
plt.ylabel("Normalized Price Index")
plt.grid(alpha=0.3)
plt.legend(title="Charizard Variant", fontsize=8)
plt.show()

# ----------------------------------------------
# 5 Correlation matrix between Charizard variants
# ----------------------------------------------
pivot = char_df.pivot_table(index="Date", columns="Card Name", values="Graded")
corr = pivot.corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, cmap="coolwarm", center=0, annot=True, fmt=".2f")
plt.title(" Correlation Between Charizard Variants (Graded Prices)")
plt.show()

# ----------------------------------------------
# 6 Prophet forecasting (optional)
# ----------------------------------------------
forecast_results = {}

for card in char_df["Card Name"].unique():
    sub = char_df[char_df["Card Name"] == card][["Date", "Graded"]].dropna()
    if len(sub) < 10:
        continue

    df_p = sub.rename(columns={"Date": "ds", "Graded": "y"})
    model = Prophet(daily_seasonality=False, yearly_seasonality=True)
    model.fit(df_p)
    future = model.make_future_dataframe(periods=365)
    forecast = model.predict(future)

    # Normalize forecast
    forecast["yhat_norm"] = forecast["yhat"] / forecast["yhat"].iloc[0] * 100
    forecast_results[card] = forecast

    plt.plot(forecast["ds"], forecast["yhat_norm"], label=card)

plt.title(" Prophet Forecast  Charizard Variants (Normalized, 12-Month Horizon)")
plt.xlabel("Date")
plt.ylabel("Predicted Price Index (Base = 100)")
plt.grid(alpha=0.3)
plt.legend(fontsize=8)
plt.show()

# ----------------------------------------------
# 7 Average Charizard Index
# ----------------------------------------------
# Combine normalized series across all variants
pivot_norm = char_norm.pivot_table(index="Date", columns="Card Name", values="Norm_Graded")
pivot_norm["Charizard_Avg_Index"] = pivot_norm.mean(axis=1)

plt.figure(figsize=(12,6))
plt.plot(pivot_norm.index, pivot_norm["Charizard_Avg_Index"], "r-", linewidth=2, label="Charizard Index")
plt.title(" Charizard Family Index (Average Normalized Graded Prices)")
plt.xlabel("Date")
plt.ylabel("Average Normalized Index (Base = 100)")
plt.grid(alpha=0.3)
plt.legend()
plt.show()

# Save summary outputs
summary.to_csv("charizard_summary.csv")
pivot_norm.to_csv("charizard_index.csv")

print(" Charizard analysis complete. Files saved: charizard_summary.csv, charizard_index.csv")
```
